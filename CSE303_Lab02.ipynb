**Question Number 1:** Find all of the numbers from 1–1000 that are divisible by 8

```python
numbers = [x for x in range(1,1000) if x%8==0]
numbers
```

**Question Number 2:** Find all of the numbers from 1–1000 that have a 6 in them 

```python
numbers = [x for x in range(1,1000) if '6' in str(x)]
numbers
```

**Question Number 3:** Count the number of spaces in a string (use string above)

```python
string = 'Number is one of the largest'
number_of_spaces = [x for x in string if x==" "]
print(len(number_of_spaces))
```

**Question Number 4:** Remove all of the vowels in a string (use string above)

```python
string = 'East West University'
remove_vowels_string = [x for x in string if x not in 'a,e,i,o,u,A,E,I,O,U'] 
print(*(remove_vowels_string))
```

**Question Number 5:** Find all of the words in a string that are less than 5 letters (use string above)

```python
string = 'East West University'
w = string.split()
words = [x for x in w if len(x)<5]
words
```

**Question Number 6:** Use a dictionary comprehension to count the length of each word in a sentence (use string above)

```python
string = 'East West University'
w = string.split()
words = [len(x) for x in w]
words
```
