```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('Boston.csv')
df.head()
```

We know if mean < Median < mode then negatively skewed or if mode < Median < mean then positively skewed.

```python
list_head = list(df)
list_head
```

```python
def hist_for_each_attri(l):
  for i in range (1,len(l),1):
    df.hist(column = [l[i]], bins = 10)
    plt.xlabel(l[i])
    plt.ylabel('count')
    plt.axvline(df[l[i]].mean(), color='red', linestyle='dashed', linewidth=2)
    plt.axvline(df[l[i]].median(), color='orange', linestyle='dashed', linewidth=2)
    if(df[l[i]].mean()>df[l[i]].median()):
      plt.title('Positively Skewed')
      #print('Positively Skewed becouse mean value is grater then median')
    else:
      plt.title('Negatively Skewed')
      #print('Negatively Skewed becouse mean value is less then median')
```

We know if mean < Median < mode then negatively skewed or if mode < Median < mean then positively skewed.

```python
hist_for_each_attri(list_head)
```

```python
def boxplot_function(l):
  
  f, (ax_box, ax_hist) = plt.subplots(2,  gridspec_kw={"height_ratios": (.30, .90)})
  
 
  sns.boxplot(l, ax=ax_box)
  sns.histplot(l, ax=ax_hist,bins = 40)
  
 
  ax_box.set(xlabel='')

  plt.show()
  q1 = np.quantile(l, 0.25)

  q3 = np.quantile(l, 0.75)
  med = np.median(l)

  iqr = q3-q1

  upper_bound = q3+(1.5*iqr)
  lower_bound = q1-(1.5*iqr)
  print('Interquartile range :',iqr)
  print('Upper bound is :',upper_bound)
  print('Lower bound is :',lower_bound)

  outliers = l[(l <= lower_bound) | (l >= upper_bound)]
  print('The following are the outliers in the boxplot:', outliers)
  print("Number of Outlers: ", len(outliers))
```

```python
boxplot_for_crim = boxplot_function(np.array(df['crim']))
```

```python
boxplot_for_zn = boxplot_function(np.array(df['zn']))
```

```python
def boxplot_for_each(l):
  for i in range (0,len(l),1):
    boxplot_function(np.array(df[l[i]]))
  return 0
```

```python
test_list = ['crim',
 'zn',
 'indus',
 'chas',
 'nox',
 'rm',
 'age',
 'dis',
 'rad',
 'tax',
 'ptratio',
 'black',
 'lstat',
 'medv']
```

```python
boxplot_for_each(test_list)
```

```python
corelationHeatmap = df[['crim','zn','indus','chas','nox','rm','age','dis','rad','tax','ptratio','black','lstat','medv']].corr()
```
